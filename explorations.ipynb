```python
import pandas as pd
import sqlite3
```

### Reading sql tables to get information about drivers and teams

```python
conn = sqlite3.connect('tour_de_france.db')
```

```python
riders = pd.read_sql_query("SELECT * FROM riders", conn)
teams = pd.read_sql_query("SELECT * FROM teams", conn)
```

```python
riders['country'] = riders['country'].apply(lambda x: x.strip('()').upper())
```

```python
# convert data to relevant data types
riders['birth'] = riders['birth'].astype('datetime64[ns]')
riders['age'] = riders['age'].astype(int)
```

```python
riders.groupby(['country'], sort=True).count()
```

```python
def split_column(df, column, new_columns):
    '''Split a column into multiple columns'''
    df[new_columns] = df[column].apply(lambda x: pd.Series(str(x).split(",")))
    df[new_columns] = df[new_columns].apply(lambda x: x.str.strip(" [''] "))
    df.drop(columns=[column], inplace=True)
```

```python
performance_cols = ['intermediate sprint', 'stages victory', 'Victories', 'Podiums']
split_column(riders, 'performance', performance_cols)

split_column(riders, 'rankings', [i for i in range(1, 22)])
```
